## Population Distribution Visualization Using World Bank Data

**Internship:** Prodigy Infotech

```
import pandas as pd

df = pd.read_csv(
    "API_SP.POP.TOTL_DS2_en_csv_v2_267553.csv",
    skiprows=4
)

print(df.head())
```

```
import matplotlib.pyplot as plt

year = "2024"

top10 = df[df["Country Name"].isin([
    "India",
    "China",
    "United States",
    "Indonesia",
    "Pakistan",
    "Nigeria",
    "Brazil",
    "Bangladesh",
    "Russian Federation",
    "Ethiopia"
])]

top10 = top10[["Country Name", year]].sort_values(by=year, ascending=False)

print(top10)

plt.figure(figsize=(12,6))
plt.bar(top10["Country Name"], top10[year])

plt.title("Top 10 Most Populous Countries (2024)")
plt.xlabel("Country")
plt.ylabel("Population")
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig("population_distribution.png")
plt.show()
```
